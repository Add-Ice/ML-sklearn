```python
from sklearn import tree
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
```

```python
wine = load_wine()
wine.data.shape
```

```python
wine.target
```

```python
import pandas as pd
```

```python
pd.concat([pd.DataFrame(wine.data),pd.DataFrame(wine.target)],axis=1)
```

```python
wine.feature_names
```

```python
wine.target_names
```

```python
# 训练集与测试卷的划分
Xtrain,Xtest,Ytrain,Ytest = train_test_split(wine.data,wine.target,test_size=0.2) # 随机划分
```

```python
# 建立模型
clf = tree.DecisionTreeClassifier(criterion='entropy')
clf = clf.fit(Xtrain,Ytrain)
score = clf.score(Xtest,Ytest) # 用于返回预测的准确度

score
```

```python
# 画树
feature_name = ['酒精','苹果酸','灰','灰的碱性','镁','总酚',
               '类黄酮','非黄烷类分类','花青素','颜色强度'
               ,'色调','od280/od315稀释葡萄酒','脯氨酸']
import graphviz
dot_data = tree.export_graphviz(clf # 所训练的模型
                               ,out_file = None # 输出路径
                               ,feature_names = feature_name
                               ,filled = True
                               ,rounded = True
                               ,class_names=['琴酒','雪莉','贝尔摩德']
                               )
graph = graphviz.Source(dot_data)
graph
```

```python
clf.feature_importances_ # 使用的特征重要性
```

```python
[*zip(feature_name,clf.feature_importances_)]#将名称与值一一对应
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy',random_state=30,splitter='random')
clf = clf.fit(Xtrain,Ytrain)
score = clf.score(Xtest,Ytest)
score
```

```python
import graphviz
dot_data = tree.export_graphviz(clf
                               ,feature_names=feature_name
                               ,filled=True
                               ,rounded=True
                               ,class_names=['琴酒','雪莉','贝尔摩德'])
graph = graphviz.Source(dot_data)
graph
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy'
                                 ,random_state=30
                                 ,max_depth=3
                                 ,min_samples_leaf=10
                                 ,min_samples_split=10)#构造模型类的实例化
clf = clf.fit(Xtrain,Ytrain)
dot_data = tree.export_graphviz(clf
                               ,feature_names=feature_name
                               ,class_names=['雪莉','威士忌','拉菲']
                               ,filled=True
                               ,rounded=True)
graph = graphviz.Source(dot_data)
graph
```

```python
clf.score(Xtrain,Ytrain)
clf.score(Xtest,Ytest)
```

```python
#采用超参数曲线来进行参数的确定
import matplotlib.pyplot as plt
test = []
test_max_depth = []
for i in range(10):
    clf = tree.DecisionTreeClassifier(max_depth=i+1
                                     ,criterion='entropy'
                                     ,random_state=30
                                     ,splitter='random')# 确定max_depth的最优值
    clf = clf.fit(Xtrain,Ytrain)
    score = clf.score(Xtest,Ytest)
    test.append(score)
    
```

```python
plt.plot(range(1,11),test,color='red',label='max_depth')
plt.legend()
plt.show()
```

```python
index = max(test)
max_index = test.index(index) + 1
max_index
```

```python
rng = np.random.RandomState(1)
x = np.sort(5 * rng.rand(80,1),axis=0)
y = np.sin(x).ravel()
len(y)
```

```python
y[::5] += 3 * (0.5 - rng.rand(16)) # 80 / 5 =16
y
```

```python
np.random.random(10)
t = np.random.random((10,1))
t.ravel().shape
```

```python
regr_1 = DecisionTreeRegressor(random_state=10,max_depth=3)
regr_2 = DecisionTreeRegressor(random_state=10,max_depth=5)
regr_1 = regr_1.fit(x,y)
regr_2 = regr_2.fit(x,y)
```

```python
x_test = np.arange(0.0,5.0,0.01)[:,np.newaxis]
x_test.shape
```

```python
y_1 = regr_1.predict(x_test)
y_2 = regr_2.predict(x_test)
(y_1.shape,y_2.shape)
```

```python
l = np.array([1,2,3,4])
l[:,np.newaxis].shape
# l[np.newaxis,:].shape
```

```python
plt.scatter(x,y,s=10,edgecolors='black',c='darkorange',label='data')
plt.plot(x_test,y_1,color='cornflowerblue',label='max_depth=3',linewidth=2)
plt.plot(x_test,y_2,color='yellowgreen',label='max_depth=5',linewidth=2)
plt.xlabel('data')
plt.ylabel('target')
plt.title('Desicion Tree Regression')
plt.legend()
plt.show()
```
